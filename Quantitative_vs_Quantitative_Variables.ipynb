```python
data = pd.read_csv(file_path)
print(data.head())
```

```python
# Define pairs of quantitative variables to analyze
pairs = [
    ("Open", "High"),
    ("Low", "Close"),
    ("High", "Volume"),
    ("Low", "Volume")
]
```

```python
# Create scatter plots for each pair
for x_var, y_var in pairs:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=numeric_data, x=x_var, y=y_var, alpha=0.6, edgecolor="black")
    plt.title(f"Scatter Plot of {x_var} vs. {y_var}", fontsize=14)
    plt.xlabel(x_var, fontsize=12)
    plt.ylabel(y_var, fontsize=12)
    plt.show()
```

```python
# Compute Covariance
covariance = numeric_data[x_var].cov(numeric_data[y_var])

# Compute Correlation
correlation = numeric_data[x_var].corr(numeric_data[y_var])
```

```python
# Print results
print(f" Covariance ({x_var} vs. {y_var}): {covariance:.2f}")
print(f" Correlation ({x_var} vs. {y_var}): {correlation:.4f}")
print("-" * 50)
```

```python
# Heatmap of Correlation Matrix
plt.figure(figsize=(10, 6))
corr_matrix = numeric_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Stock Variables", fontsize=14)
plt.show()
```

```python

# Ensure only numeric columns are used
numeric_data = data.select_dtypes(include=["number"]).replace([np.inf, -np.inf], np.nan).dropna()

#  Create new variables for better trading insights
data["Price Change"] = data["High"] - data["Low"]  # Measures daily volatility
data["Daily Return %"] = ((data["Close"] - data["Open"]) / data["Open"]) * 100  # % change in price
data["Volatility %"] = ((data["High"] - data["Low"]) / data["Open"]) * 100  # % range between High and Low
data["Log Volume"] = np.log1p(data["Volume"])  # Log scale for better visualization of volume

# Set seaborn style
sns.set_style("whitegrid")

#  New Visualization Set: Trading Insights

# 1 Scatter Plot: Price Change vs. Volume (Are large price swings linked to trading activity?)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Price Change", y="Log Volume", alpha=0.6, edgecolor="black")
plt.title("Price Change vs. Trading Volume", fontsize=14)
plt.xlabel("Daily Price Change ($)", fontsize=12)
plt.ylabel("Log of Trading Volume", fontsize=12)
plt.show()

# 2 Scatter Plot: Daily Return % vs. Volume (Does heavy trading lead to big returns?)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Daily Return %", y="Log Volume", alpha=0.6, edgecolor="black")
plt.title("Daily Return % vs. Trading Volume", fontsize=14)
plt.xlabel("Daily Return (%)", fontsize=12)
plt.ylabel("Log of Trading Volume", fontsize=12)
plt.show()

# 3 Scatter Plot: Volatility % vs. Volume (Is volatility higher on high-volume days?)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Volatility %", y="Log Volume", alpha=0.6, edgecolor="black")
plt.title("Volatility % vs. Trading Volume", fontsize=14)
plt.xlabel("Volatility (%)", fontsize=12)
plt.ylabel("Log of Trading Volume", fontsize=12)
plt.show()

# 4 Scatter Plot: Close Price vs. Daily Return % (Are higher-priced stocks more volatile?)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Close", y="Daily Return %", alpha=0.6, edgecolor="black")
plt.title("Stock Price vs. Daily Return %", fontsize=14)
plt.xlabel("Closing Price ($)", fontsize=12)
plt.ylabel("Daily Return (%)", fontsize=12)
plt.show()

# 5 Heatmap: New Correlation Matrix (For deeper relationships)
plt.figure(figsize=(10, 6))
corr_matrix = data[["Price Change", "Daily Return %", "Volatility %", "Log Volume", "Close"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Trading Metrics", fontsize=14)
plt.show()

# 6 Pairplot: How do these new variables interact?
sns.pairplot(data[["Price Change", "Daily Return %", "Volatility %", "Log Volume", "Close"]], diag_kind="kde", markers="o", corner=True)
plt.suptitle("Pairwise Comparison of New Trading Metrics", y=1.02, fontsize=14)
plt.show()
```

### ** Summary of the Trading Visualizations**

---

## **1 Scatter Plot: Price Change vs. Trading Volume**  
 **What it shows:**  
- This plot examines whether **larger price swings (High - Low) are linked to increased trading volume**.  

 **Expected Insights:**  
- If high volume is associated with large price swings, it suggests **higher market activity on volatile days**.  
- A **strong correlation** could indicate **panic buying/selling** or institutional trades driving volatility.  
- If theres **no clear pattern**, it suggests that **price swings happen independently of trading volume**.  

---

## **2 Scatter Plot: Daily Return % vs. Trading Volume**  
 **What it shows:**  
- Investigates whether **higher trading volume leads to bigger percentage returns**.  
- **Daily Return % = (Close - Open) / Open  100**  

 **Expected Insights:**  
- **Positive correlation** could mean **big trading days drive stock gains**.  
- **Negative correlation** could indicate **panic selling or price manipulation**.  
- **No correlation?** Suggests stock prices **fluctuate independently of trading volume**.  

---

## **3 Scatter Plot: Volatility % vs. Trading Volume**  
 **What it shows:**  
- Measures whether **high trading volume is linked to high price volatility**.  
- **Volatility % = (High - Low) / Open  100**  

 **Expected Insights:**  
- A **strong correlation** suggests that **high volume days are more unpredictable**.  
- A **flat or weak correlation** means **price stability is not necessarily affected by volume spikes**.  

---

## **4 Scatter Plot: Closing Price vs. Daily Return %**  
 **What it shows:**  
- Compares whether **higher-priced stocks experience more volatility in daily returns**.  

 **Expected Insights:**  
- **If high-priced stocks have low returns**, they are **more stable investments**.  
- **If high-priced stocks show high returns**, they could be **more speculative and volatile**.  
- **Low-priced stocks with large return swings?** They may be **risky and unpredictable**.  

---

## **5 Heatmap: Correlation Matrix of Trading Metrics**  
 **What it shows:**  
- A heatmap **summarizes the relationships between all new trading variables**.  
- **Bright red (near +1)** = Strong positive correlation.  
- **Bright blue (near -1)** = Strong negative correlation.  

 **Expected Insights:**  
- **Strong correlation** between **Price Change, Volatility %, and Volume** could indicate **high volatility on active trading days**.  
- **Low correlation between Daily Return % and Volume** suggests that **trading activity does not always lead to big price movements**.  

---

## **6 Pairplot: Pairwise Comparisons of Trading Metrics**  
 **What it shows:**  
- A **comprehensive look at all trading variables** side by side.  
- Scatter plots for relationships, **density plots for individual distributions**.  

 **Expected Insights:**  
- **Clear linear trends** suggest strong relationships.  
- **Scattered points** indicate **weak or no correlation**.  
- **Outliers** can highlight **unusual market behaviors** (e.g., one day had extreme volume or returns).  

---

### ** Key Takeaways**
 **Price swings and volume may be correlated**, but not always.  
 **Stock price levels affect return stability** (higher prices may reduce volatility).  
 **High trading volume doesnt always mean high returns** (depends on market sentiment).  
 **Heatmap helps quickly identify strong relationships**.  
 **Pairplot offers an overall picture of how all metrics interact**.
